The goal of this assignment is to compare the trajectories of several different autonomous second order linear systems of the type
$$
    \begin{bmatrix}
        \dot{x}_1 \\
        \dot{x}_2 \\
    \end{bmatrix}
     =
     A
    \begin{bmatrix}
        x_1 \\
        x_2 \\
    \end{bmatrix}
$$
and see how things may look like in different situations and under various changes of bases. Throughout this assignment, you are intended to use the provided python functions such as plot_phase_portrait. You will NOT need to modify these functions.

```python
####################################################################
# It is not necessary to change/modify any of the code in this block
####################################################################

import numpy as np
import matplotlib.pyplot as plt

def plot_phase_portrait(A, title, x_range=(-10, 10), y_range=(-10, 10), num_grid_points=20):
    """
    Plots the phase portrait for a system with matrix A.
    
    Parameters:
    - A: 2x2 numpy array, system matrix
    - title: string, title of the plot
    - x_range: tuple, range for x-axis
    - y_range: tuple, range for y-axis
    - num_grid_points: int, number of grid points in each direction
    """
    x1 = np.linspace(x_range[0], x_range[1], num_grid_points)
    x2 = np.linspace(y_range[0], y_range[1], num_grid_points)
    
    X1, X2 = np.meshgrid(x1, x2)
    U = A[0, 0] * X1 + A[0, 1] * X2
    V = A[1, 0] * X1 + A[1, 1] * X2
    
    plt.figure(figsize=(6, 6))
    plt.streamplot(X1, X2, U, V, color='b', linewidth=1)
    plt.title(title)
    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()
```

Generate a random $2 \times 2$ matrix, in the sense of generating randomly its four elements, and check its determinant. If it is bigger than $0.0001$ then use that matrix for the steps below, otherwise re-generate another one.

Use the generated matrix as a change of basis, and generate the corresponding new phase portraits for all the various $T A_{ij} T^{-1}$'s, with the $A_{ij}$ being the matrices you saw in the previous questions.

If you were to select randomly one of such 'changed-basis' phase portraits, would you be able to correctly identify from which $A_{ij}$ it would come from?

If so for all the potential choices, then good - you are done! If not, then re-start this question over, again and again until you feel confident on identifying from which case a generic phase portrait comes from.

Use the provided functions generate_random_matrix and plot_transformed_phase_portraits to answer the question

```python
####################################################################
# It is not necessary to change/modify any of the code in this block
####################################################################

def generate_random_matrix(threshold=0.0001):
    """Generates a random 2x2 matrix with a determinant larger than the threshold."""
    while True:
        T = np.random.rand(2, 2)  # Generate random 2x2 matrix
        det = np.linalg.det(T)
        if abs(det) > threshold:
            return T

def plot_transformed_phase_portraits(T, matrices, labels):
    """Plots the phase portraits for the transformed matrices T A T^{-1}."""
    T_inv = np.linalg.inv(T)
    for A, label in zip(matrices, labels):
        A_transformed = T @ A @ T_inv
        plot_phase_portrait(A_transformed, f'Phase Portrait for Transformed {label}')

# Generate a random 2x2 matrix T with a sufficient determinant
T = generate_random_matrix()
```

```python
#############
# Example use
#############

A1 = np.array([[-1, 0], [0, -1]])
A2 = np.array([[1, 0], [0, 1]])

matrices = [A1, A2]
labels = ['A1', 'A2']

plot_transformed_phase_portraits(T, matrices, labels)
```
